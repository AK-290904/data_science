#Operators

```python
#Arithmetic Operators
print(5+6)

print(5/2)

#integer division-->take integral part only
print(5//2)



#modulo
print(5%2)

#exponential
print(5**2)
```


Relational operator



```python
print(4<5)
print(4<=4)
```

Logical Operator

```python
print(1 and 0)

print(1 or 0)

print(not 1)
```

Bitwise Operator-> operate on binary



```python
#bitwise and
print(2 & 3)

#bitwise or
print(2 | 3)

#bitwise xor-->same gives 0 and diff gives 1
print(2 ^ 3)


#bitwise not
print(~3)

#left shift
print(4 >> 2)

#right shift
print(5 << 2)
```

Assignment Operator

```python
a=2
a+=2  #a=a+2
print(a)


#in python a++ and ++a is not valid
```

Membership Operator

```python
#in/not in
print('D' not in 'Delhi')
print(1 in [1,2,3,4,5,6])
```

#If-else

```python
#login program and check valid identification
#Assume: email->abc@gmail.com
#pass->1234

email=input('Enter email ')
password=input('Enter Password ')

if email=='abc@gmail.com' and password=='1234':
  print('Welcome')
else:
  print('Not valid')
```

```python
#if-elif-else

email=input('Enter email ')
password=input('Enter Password ')

if email=='abc@gmail.com' and password=='1234':
  print('Welcome')
elif email=='abc@gmail.com' and password!='1234':
  print('Incorrec Pass')
  password=input('Enter pass again')
  if password=='1234':
    print('Welcome')
  else:
    print('Not valid')
else:
  print('Not valid')
```

```python
#menu driven Calculator

fnum=int(input('Enter 1st no'))
snum=int(input('Enter 2nd no'))

op=input('Enter operator')

if op=='+':
  print(fnum+snum)
elif op=='-':
  print(fnum-snum)
elif op=='*':
  print(fnum*snum)
elif op=='/':
  print(fnum/snum)
else:
  print('Invalid operator')
```

#Loops in Python


```python
#while loop

#Printing table

number=int(input('Enter no '))

i=1
while i<11:
  print(number*i)
  i+=1
```

```python
 #while loop with else

x=1
while x<3:
  print(x)
  x+=1
else:
  print('Limit crossed')
```

```python
#Guessing Game

#generate a random integer between 1 to 100

import random
jackpot=random.randint(1,100)

guess=int(input('Guess no'))
counter=1
while guess!=jackpot:
  if guess<jackpot:
    print('Guess Higher')
  else:
    print('Guess Lower')

  guess=int(input('Guess no'))
  counter+=1
else:
  print('Correct Guess')
  print('Attempts',counter)
```
