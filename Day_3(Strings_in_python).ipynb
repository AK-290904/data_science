#Strings
In python,strings are a sequence of unicode characters.

Note->

  unicode was used in place of ASCII as it store 8 bits only ie 2^8 combinations can be represensted for characters but as time passes need for representing ch increases like we use emoji so unicode hold 16 bits hence 2^16 combinations.

#Creating Strings

```python
#we can create strings  using single double and triple inverted commas
s='hello'
s="hello"
#triple used for multiline strings
s='''hello'''
s="""hello"""
print(s)
```

```python
#We can also create strings using type convertion
s=str('hello')
print(s)
```

```python
#we have both single and double inverted commas because when we use single ' inside sentence so compiler dont get confuse we can use " "
s="it's raining"
print (s)
```

#Accessing Substring from a string

```python
#Positive indexing
s='hello world'
print(s[0])
```

```python
#Negtive indexing->w/o length we can extract last char in string
s='hello world'
print(s[-1])
```

```python
#Slicing->more than one characters from string
s='hello world'
print(s[0:5])#exclude end
print(s[:])
print(s[0:6:2])#we can give step also as 3rd arg
```

```python
s='hello world'
print(s[6:0:-1])
```

```python
#to reverse a string
s='hello world'
print(s[::-1])
```

```python
s='hello world'
print(s[-1:-6:-1])#when we use neg indexing 1st arg is greater than 2nd
```

#Editing and Deleting in Strings

```python
s='hello world'
s[0]='H' #it will give error 'str' object does not support item assignment
#because in python strings are immutable
```

```python
from os import spawnl
#To delete a string
s='hello world'
del s
print(s)#it will give error name 's' is not defined as s is deleted in above line
```

```python
#whole str can be deleted but not a part using slicing as str are immutable
s='hello world'
del s[-1:5:2]
print(s)#error 'str' object does not support item deletion
```
