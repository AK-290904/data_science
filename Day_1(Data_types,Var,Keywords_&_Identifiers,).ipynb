#Type Conversion

```python
#Implicit Conversion
print(5+5.6)

#but not always works implicit
#eg:   print(1+'a')-->ERROR
```

```python
#Explicit Type Conversion
#str->int
a=int('4')
type(a)

```

```python
#int to str
a=str(5)
type(a)
```

```python
#int to float
a=float(5)
type(a)
```

```python
#Python type conversion do not chnage the exsisting variable it create new object

fnum=input('1st no ')
#fnum=int(input('1st no'))
snum=input('2nd no ')
print(type(fnum),type(snum))
result=int(fnum)+int(snum)
print('Result ',result)
print(type(result))

print(type(fnum))

```

#Literals

```python
a=0b10101 #Binary Literals
b=100 #Decimal Literals
c=0o310  #Octal Literals
d=0x12c  #Hexadecimal Literals

#Float Literals
f1=10.5
f2=1.5e2
f3=1.5e-3

#Complex Literals
x=3.14j

print(a,b,c,d)
print(f1,f2,f3)
print(x,x.imag,x.real)
```
